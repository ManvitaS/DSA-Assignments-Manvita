#### 1. Read the dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_excel('iris.xls')
data
```

#### 2. Display columns in dataset

```python
data.columns
```

#### 3. Mean of each column of the dataset.

```python
print('Mean of column SL:', np.mean(data['SL']))
print('Mean of column SW:', np.mean(data['SW']))
print('Mean of column PL:', np.mean(data['PL']))
print('Mean of column PW:', np.mean(data['PW']))
```

#### 4. Check for null values in the dataset

```python
pd.isnull(data)
```

#### 5. Perform meaningful visualizations using the dataset

###### Histogram

```python
plt.figure()
plt.hist(data['SL'],rwidth=0.8,color='purple',density='True',cumulative='True')
plt.title('Distribution of SL')
plt.xticks()
plt.yticks()
```

```python
data[['SL','SW']].plot.kde()
```

###### Scatter Plot

```python
plt.figure(figsize=(8,4))
plt.scatter(data['PL'],data['PW'],s=12,c='red',marker='*')
plt.title('Plot of PL vs PW',fontsize=14)
plt.xlabel('PL')
plt.ylabel('PW')
```

###### Box Plot

```python
sns.boxplot(x='Classification',y='SL',data=data)
```

###### Strip Plot

```python
sns.stripplot(x = data['SW'])
```

###### Swarm Plot

```python
sns.swarmplot(x='Classification',y='SW',data=data)
```

###### Violin Plot

```python
sns.violinplot(x='Classification',y='SL',data=data)
```

###### Count Plot

```python
sns.countplot(x=data['Classification'])
```

###### Pair Plot

```python
slice_data = data[['SL','SW','PL','PW']]
sns.pairplot(slice_data)
```
