### Case study on Presentation of data 

Do the following operations on ‘cars_data’ dataset:
    
1. Read the dataset to the python environment.
2. Check for the null values present in the dataset.
3. Plot a bar graph of male vs female buyers participated in the sales.
4. Find the top 5 cars based on their sales price.
5. Find the least 5 cars based on their Resell price.

#### 1. Read the dataset to the python environment.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

cars_data = pd.read_csv('cars_data.csv')
cars_data
```

#### 2. Check for the null values present in the dataset.

```python
cars_data.isnull().sum()
```

#### Insight :-
There are no null values present in the given dataset.

#### 3. Plot a bar graph of male vs female buyers participated in the sales.

```python
cars_data1 = cars_data.loc[cars_data["Buyer Gender"] == "Male"].count()
print('No.of male buyers :-', cars_data1["Buyer Gender"])
cars_data2 = cars_data.loc[cars_data["Buyer Gender"] == "Female"].count()
print('No.of female buyers :-', cars_data2["Buyer Gender"])
sns.countplot(x ='Buyer Gender', data = cars_data)
plt.ylim(4500, 5500)
plt.title("Male Vs Female Buyers participated in the sales", fontsize = 12)
plt.show()
```

#### Insight :- 
The number of female buyers are slightly more than the male buyers who participated in the sales.

#### 4.Find the top 5 cars based on their sales price.

```python
cars_data.nlargest(5, ['Sale Price'])
```

#### Insight :- 
The top 5 cars having the highest sales price are :- Volkswagen Jetta III, Audi A6, Lexus LX, Mazda MX-6 and Lexus ES

#### 5.Find the least 5 cars based on their Resell price

```python
cars_data.nsmallest(5,['Resell Price'])
```

#### Insight :-
5 cars having the lowest Resell price are :- Chevrolet 1500, GMC Savana 1500, Jaguar XF, Mistubishi 30000GT and Infiniti G37
